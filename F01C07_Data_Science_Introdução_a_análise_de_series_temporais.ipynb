# <font color=navyblue> AULA 02 - Decomposição e autocorrelação



Resumo:
* criação de novas colunas por decomposição usando a função diff()
* construção de função para agilizar o plot dos gráficos de linha já configurados
* configuração de subplots para alocar três gráficos no mesmo plot
* construção de função para agilizar os subplots com três gráficos
* plot de gráficos de autocorrelação usando as colunas de alucar

## 1) Decompondo as vendas

```
## pelo gráfico da aula passada, é notável o crescimento das vendas ao longo do tempo
## que outras informações eu posso extrair desse gráfico?

alucar.head()
```

```
## façamos o seguinte: criemos uma nova coluna com a diferença entre as vendas de dois meses subsequentes
## ou seja, a variação das vendas de um mês para outro

## a função series.diff() faz isso automaticamente --> https://pandas.pydata.org/docs/reference/api/pandas.Series.diff.html

alucar['aumento'] = alucar['vendas'].diff()
alucar.head()
```

```
## plotemos um gráfico com os valores da coluna aumento

## configurando a paleta de cores do gráfico --> https://seaborn.pydata.org/generated/seaborn.set_palette.html
sns.set_palette('Accent')  


## adicionando grid ao gráfico --> https://seaborn.pydata.org/generated/seaborn.set_style.html
sns.set_style('darkgrid')


## atribuindo o gráfico a uma variável, ax, retirando o título <matplotlib.axes._subplots.AxesSubplot at 0x7f12d87a4ad0>
## tal nome representava o local de armazenamento do gráfico
ax = sns.lineplot(x='mes', y='aumento', data=alucar)


## aumentando o tamanho do gráfico --> https://matplotlib.org/3.3.4/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure.set_size_inches
## dessa vez é uma configuração do matplotlib
ax.figure.set_size_inches(12,6)

## adicionando título ao gráfico, alinhando à esquerda e ajustando o tamanho da fonte --> https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.set_title.html
ax.set_title('Variação das vendas Alucar de 2017 a 2018', loc='left', fontsize=18)

## modificando texto e fonte dos rótulos dos eixos --> https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html
ax.set_xlabel('Meses', fontsize=14)
ax.set_ylabel('Vendas', fontsize=14)


## plt.xticks(rotation=90) se eu quiser rotacionar os rótulos do eixo X


plt.show()
```

```
## para otimizar esse processo de criação de gráfico, podemos transformar o código da célula anterior em uma função

def plotar(titulo, xlabel, ylabel, x, y, dataset):
  sns.set_palette('Accent')  
  sns.set_style('darkgrid')
  ax = sns.lineplot(x=x, y=y, data=dataset)
  ax.figure.set_size_inches(12,6)
  ax.set_title(titulo, loc='left', fontsize=18)
  ax.set_xlabel(xlabel, fontsize=14)
  ax.set_ylabel(ylabel, fontsize=14)
  plt.show()
```

```
## vamos testar a função

plotar('Variação das vendas Alucar de 2017 a 2018', 'Meses', 'Variação das vendas', 'mes', 'aumento', alucar)
```

```
## o quão rápido nossas vendas estão crescendo? qual é o ritmo desse crescimento?
## façamos novamente o diff(), dessa vez na coluna 'aumento'
## criaremos assim a coluna 'aceleracao'

alucar['aceleracao'] = alucar.aumento.diff()
alucar.head()
```

```
plotar('Aceleração das vendas Alucar de 2017 a 2018', 'Meses', 'Aceleração das vendas', 'mes', 'aceleracao', alucar)
```

## 2) Observação e tendência

```
## fizemos três gráficos com informações importantes
## infelizmente, cada um está numa parte do notebook
## seria mais fácil se todos estivesse mais próximos, mesmo que em plots diferentes

## https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html

## criemos um espaço que possa abranger os três gráficos


plt.figure(figsize=(20,25))                                                             ## criando um espaço para colocar os três gráficos

ax = plt.subplot(3,1,1)                                                                 ## 3 linhas, 1 coluna, ocupe o espaço 1
ax.set_title('Análise de vendas da Alucar de 2017 a 2018', fontsize=18, loc='left')     ## título, tamanho da fonte e alinhamento
sns.lineplot(x='mes', y='vendas', data=alucar)                                          ## primeiro subplot

ax = plt.subplot(3,1,2)                                                                 ## 3 linhas, 1 coluna, ocupe o espaço 2
ax.set_title('Título 2', fontsize=18, loc='left')                                       ## título, tamanho da fonte e alinhamento
sns.lineplot(x='mes', y='aumento', data=alucar)                                         ## segundo subplot

ax = plt.subplot(3,1,3)                                                                 ## 3 linhas, 1 coluna, ocupe o espaço 3
ax.set_title('Título 3', fontsize=18, loc='left')                                       ## título, tamanho da fonte e alinhamento
sns.lineplot(x='mes', y='aceleracao', data=alucar)                                      ## terceiro subplot

plt.show()



## obs.: se quiser configurar eixos, rótulos, tamanhos, títulos etc, basta colocar os parâmetros abaixo de cada subplot
```

```
## transformaremos esse plot triplo numa função também

def plot_comparacao(titulo, x, y1, y2, y3, dataset):
  plt.figure(figsize=(20,25))                                    ## criando um espaço para colocar os três gráficos

  ax = plt.subplot(3,1,1)                                        ## 3 linhas, 1 coluna, ocupe o espaço 1
  ax.set_title(titulo, fontsize=18, loc='left')                  ## título, tamanho da fonte e alinhamento
  sns.lineplot(x=x, y=y1, data=dataset)                          ## primeiro subplot

  plt.subplot(3,1,2)                                             ## 3 linhas, 1 coluna, ocupe o espaço 2
  sns.lineplot(x=x, y=y2, data=dataset)                          ## segundo subplot

  plt.subplot(3,1,3)                                             ## 3 linhas, 1 coluna, ocupe o espaço 3
  sns.lineplot(x=x, y=y3, data=dataset)                          ## terceiro subplot

  plt.show()
```

```
## testando a função plot_comparacao

plot_comparacao('Análise de vendas da Alucar de 2017 a 2018', 'mes', 'vendas', 'aumento', 'aceleracao', alucar)
##               titulo                                        x      y1        y2         y3           dataset


## funcionou certinho
```

## 3) Autocorrelação

```
## os aumentos desse mês estão ligados com os do mês passado?
## para isso, utilizaremos a função de autocorrelação
  ## é uma medida de correlação entre as observações de uma série temporal, comparando os valores com eles mesmos
## valores de correlação entre -1 e 1
```

Lag 0: 1 compara com 1, 2 compara com 2 etc


lag0.png

Lag 1: 1 compara com 2, 2 compara com 3 etc

lag1.png

```
## usaremos a seguinte biblioteca

from pandas.plotting import autocorrelation_plot

## https://pandas.pydata.org/docs/reference/api/pandas.plotting.autocorrelation_plot.html
```

```
## autocorrelação da coluna 'vendas'

plt.figure(figsize=(12,6))
ax = autocorrelation_plot(alucar['vendas'])
ax.set_title('Correlação das vendas', fontsize=18, loc='left')
plt.show()
```

```
## autocorrelação da coluna 'aumento'

plt.figure(figsize=(12,6))
ax = autocorrelation_plot(alucar['aumento'][1:])           ## o primeiro valor é um NaN; por isso, devemos começar do segundo valor
ax.set_title('Correlação do aumento das vendas', fontsize=18, loc='left')
plt.show()
```

```
## autocorrelação da coluna 'aceleracao'

plt.figure(figsize=(12,6))
ax = autocorrelation_plot(alucar['aceleracao'][2:])           ## os dois primeiros valores são NaN; por isso, devemos começar do terceiro valor
ax.set_title('Correlação da aceleração do aumento das vendas', fontsize=18, loc='left')
plt.show()
```

## 3) Investigando a sazonalidade

```
## a chocolura possui muitas lojas ao redor do brasil
## e uma delas chamou a atenção pela quantidade de vendas diárias em outubro e novembro de 2018
## o que aconteceu?


vendas_por_dia = pd.read_csv('vendas_por_dia.csv')
vendas_por_dia.head()
```

```
## dados nulos? não

vendas_por_dia.isna().sum()
```

```
## coluna 'mes' deve ficar no formato datetime

vendas_por_dia['dia'] = pd.to_datetime(vendas_por_dia['dia'])
vendas_por_dia.dtypes
```

```
## criando colunas com diff()

vendas_por_dia['aumento'] = vendas_por_dia.vendas.diff()
vendas_por_dia['aceleracao'] = vendas_por_dia.aumento.diff()
vendas_por_dia.head()
```

```
## representação gráfica das três colunas

plot_comparacao('Análise de vendas diárias de chocolate em outubro e novembro de 2018', 'dia', 'vendas', 'aumento', 'aceleracao', vendas_por_dia)

## mais um padrão sazonal, dessa vez num período menor
## o que causa essa sazonalidade?
```

```
## para uma análise mais completa, criemos uma coluna com os dias da semana de cada um dos dias analisados
## sabemos que o primeiro dia do dataset é uma segunda feira

## existe uma função para determinar automaticamente o dia da semana para cada data
## https://pandas.pydata.org/docs/reference/api/pandas.Series.dt.day_name.html

vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.day_name()
vendas_por_dia.head()
```

```
## vamos traduzir esses dias da semana? usaremos series.map()
## pandas.Series.map https://pandas.pydata.org/docs/reference/api/pandas.Series.map.html

vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map({'Monday':'Segunda', 'Tuesday':'Terça', 'Wednesday':'Quarta',
                                     'Thursday':'Quinta', 'Friday':'Sexta', 'Saturday':'Sábado',
                                     'Sunday':'Domingo'})
```

```
## vejamos o que aconteceu nas duas primeiras semanas

vendas_por_dia.head(15)

## forte aumento das vendas nas segundas
## vende pouco nos finais de semana e nos feriados
```

```
## agrupemos as informações por dia da semana para avaliar sua influência nas vendas diárias
## pandas.DataFrame.groupby https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

vendas_por_dia.groupby(by='dia_da_semana').mean().round()

## dias mais fracos: sábado e domingo
## as vendas são mais altas nos dias úteis
```

* criação de três gráficos alocados no mesmo plot, utilizando a função criada na aula 02

## 1) Ruído de uma time series

```
## vamos importar mais um dataset, dessa vez com dados sobre venda de celulares pela empresa alucel

alucel = pd.read_csv('alucel.csv')
alucel.head()
```

```
## dados nulos? não

alucel.isna().sum()
```

```
## coluna 'mes' deve ficar no formato datetime

alucel['dia'] = pd.to_datetime(alucel['dia'])
alucel.dtypes
```

```
## criando colunas com diff()

alucel['aumento'] = alucel.vendas.diff()
alucel['aceleracao'] = alucel.aumento.diff()
alucel.head()
```

```
plot_comparacao('Análise de vendas da Alucel - Outubro/Novembro de 2018', 'dia', 'vendas', 'aumento', 'aceleracao', alucel)

## nota-se bastante ruído no gráfico de vendas, fica difícil analisar a tendência e a sazonalidade nas vendas
```

## 2) Média móvel

```
## como minimizar esses ruídos?
## usaremos a técnica de média móvel para suavizar esse gráfico e reduzir esse ruído

## https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rolling.html

alucel['media_movel'] = alucel['vendas'].rolling(7).mean()        ## média móvel de 7 dias (só começa a contar a partir do sétimo dia, considerando a média dos seis dias anteriores)
alucel.head(8)
```

```
plotar('Análise das vendas com média móvel de 7 dias', 'Tempo', 'Média móvel de vendas (7 dias)', 'dia', 'media_movel', alucel)

## já reduziu um pouco o ruído
```

```
alucel['media_movel_21'] = alucel['vendas'].rolling(21).mean()            ## média móvel de 21 dias
plotar('Análise das vendas com média móvel de 21 dias', 'Tempo', 'Média móvel de vendas (21 dias)', 'dia', 'media_movel_21', alucel)

## o ruído reduziu bastante usando a média móvel de 21 dias, mas há perda de informação
```

```
## pra terminar, plotemos os três gráficos (vendas, média móvel de 7 dias e média móvel de 21 dias)

plot_comparacao('Análise de vendas da Alucel -- Outubro e Novembro de 2018', 'dia', 'vendas', 'media_movel', 'media_movel_21', alucel)
```

# <font color=navyblue> AULA 01 - Importando e traduzindo os dados

Documentação dos comandos da aula

## 1) Importando os dados

```
## os dados utilizados no curso são os fornecidos pela própria biblioteca, com algumas modificações para fins didáticos
## faremos uma análise das gorjetas dadas pelos clientes em um dado restaurante
## para importá-los: Arquivos -> Fazer upload
## para utilizá-los, importaremos o pandas para analisar os dados apresentados

import pandas as pd
dados = pd.read_csv('tips.csv')
dados

## o df possui o índice + 5 colunas:
  ## total_bill -> total da conta
  ## tip -> valor da gorjeta
  ## dessert -> o cliente pediu sobremesa?
  ## day -> dia da semana
  ## time -> período do dia no qual foi feito o pedido
  ## size -> quantas pessoas estavam na mesa
```

```
## as colunas estão em inglês; seria uma boa traduzir para português, né?
```

## 2) Traduzindo colunas e campos

```
## vejamos quais colunas existem no nosso DataFrame

dados.columns
```

```
## criemos um dicionário para renomear as colunas usando o dados.rename

renomear = {
    'total_bill':'valor_da_conta', 
    'tip':'gorjeta', 
    'dessert':'sobremesa', 
    'day':'dia_da_semana', 
    'time':'hora_do_dia', 
    'size':'total_de_pessoas'
}

renomear
```

```
## feito o dicionário, precisamos renomear as colunas do nosso df
## usaremos dados.rename(columns = renomear) e atribuiremos a uma nova variável
## documentação: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html

gorjetas = dados.rename(columns = renomear)
gorjetas
```

```
## os rótulos das colunas já foram traduzidos, mas nota-se que o conteúdo do DataFrame ainda está em inglês
## como traduzir isso?

## vejamos a coluna gorjeta.sobremesa
## para listar os valores únicos da coluna, usaremos Series.unique()

gorjetas.sobremesa.unique()
```

```
## criaremos um dicionário para renomar os dados da coluna gorjeta.sobremesa

sim_nao = {
    'No':'Não',
    'Yes':'Sim'
}
```

```
## usemos o comando Series.map() para alterar a Series
## documentação: https://pandas.pydata.org/docs/reference/api/pandas.Series.map.html

gorjetas['sobremesa'] = gorjetas['sobremesa'].map(sim_nao)
gorjetas.head(5)

## o Series.map() mapeia a Series e procura, com base no dicionário utilizado como parâmetro, os campos a serem substituídos
## neste caso, o comando seguiu o dicionário sim_nao
## todo 'No' deve ser substituído por 'Não', todo 'Yes' deve ser substituído por 'Sim'
## se não houver nenhum dos dois casos, o valor será substituído por NaN
```

```
## façamos o mesmo para os dias da semana

gorjetas.dia_da_semana.unique()
```

```
dias = {
    'Sun':'Domingo', 
    'Sat':'Sábado', 
    'Thur':'Quinta',
     'Fri':'Sexta'
}
```

```
gorjetas.dia_da_semana = gorjetas.dia_da_semana.map(dias)
gorjetas.head(5)
```

```
## façamos mais uma vez para a hora do dia

gorjetas.hora_do_dia.unique()
```

```
hora = {
    'Dinner':'Jantar',
    'Lunch':'Almoço'
}
```

```
gorjetas.hora_do_dia = gorjetas.hora_do_dia.map(hora)
gorjetas.head()
```

```
## podemos verificar a relação entre o valor da conta e a porcentagem da gorjeta em relação ao valor total
## nota-se uma queda mais acentuada no gráfico daqueles que pediram sobremesa, 

sns.lmplot(x='valor_da_conta', y='porcentagem', data=gorjetas, col='sobremesa', hue='sobremesa')
```

**Visualmente, existe uma diferença no valor da gorjeta entre aqueles que pediram sobremesa e os que não pediram sobremesa**

## 3) Teste de hipótese


### H<sup>null</sup> (Hipótese nula)
A distribuição da taxa da gorjeta é a mesma nos dois grupos

### H<sup>alt</sup> (Hipótese alternativa)
A distribuição da taxa da gorjeta não é a mesma nos dois grupos

```
## para consolidar a análise, faremos um teste de hipótese com o objetivo de avaliar se pedir sobremesa afeta o valor da gorjeta
## descobriremos se a população geral que consome a sobremesa e a que não consome se comportam igual ou diferentemente da amostra
```

```
## para realizarmos nosso teste de hipótese, importaremos uma nova biblioteca (ranksums)
## documentação: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ranksums.html

from scipy.stats import ranksums
```

```
## traremos o valor da porcentagem da gorjeta de todas as pessoas que pediram sobremesa a partir da montagem de uma query

sobremesa = gorjetas.query('sobremesa == "Sim"')['porcentagem']    ## sobremesa = Sim
sobremesa
```

```
## query com os valores da porcentagem da gorjeta daqueles que não pediram sobremesa

sem_sobremesa = gorjetas.query('sobremesa == "Não"')['porcentagem']    ## sobremesa = Não
sem_sobremesa
```

```
## para efetivar o teste de hipótese, usaremos o ranksums

r = ranksums(sobremesa, sem_sobremesa)
r

## p-value como 0,52 mostra que a diferença entre os percentuais de gorjeta é matematicamente insignificante
## como p-value é maior que 0,05, a hipótese alternativa deve ser rejeitada

## dessa forma, a hipótese nula tende a ser a mais correta
## portanto, podemos dizer que a distribuição do percentual de gorjeta é a mesma nos dois grupos
```

```
print(f'O valor do p-value é {r.pvalue:.2f}')
```

### c) Análise 3 - Dia da semana e gorjeta

## 2) Análise descritiva

```
## agora, façamos análises descritivas em relação ao valor da conta, porcentagem, média de gorjetas e etc

## qual é a média das gorjetas do df?

media_gorjeta = gorjetas.gorjeta.mean()
print(f'A média geral das gorjetas é de US$ {media_gorjeta:.2f}')
```

```
## e qual é a média das gorjetas para cada dia da semana?
## usemos o groupby
## documentação: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html

gorjetas.groupby('dia_da_semana').mean()[['valor_da_conta', 'gorjeta', 'porcentagem']]   ## double colchete no final para listar as colunas desejadas no df
```

```
## nota-se que a média da gorjeta no domingo é maior do que as dos outros dias
## mas quantas mesas foram atendidos no restaurante em cada dia da semana?

gorjetas.dia_da_semana.value_counts()
```

```
## mas nem toda mesa tem a mesma quantidade de pessoas
## quantos clientes foram atendidos em cada dia da semana?

gorjetas.groupby('dia_da_semana').sum()['total_de_pessoas']
```

## 3) Teste de hipótese

### H<sup>null</sup> (Hipótese nula)  p-value > 0,05
A distribuição do valor da conta é a mesma nos dois dias

### H<sup>alt</sup> (Hipótese alternativa)  p-value < 0,05
A distribuição do valor da conta não é a mesma nos dois dias

```
## observamos que a média do valor da gorjeta no domingo é maior do que a do sábado
## porém, a quantidade de pessoas no restaurante é maior no sábado do que no domingo

## a questão é: os valores das contas nos dois dias são diferentes também?
```

```
## hora de fazer mais um teste de hipótese

## façamos duas queries separando apenas os valores de contas do sábado e do domingo

## SÁBADO
## outra opção: gorjetas[gorjetas.dia_da_semana == 'Sábado'].valor_da_conta

valor_sabado = gorjetas.query('dia_da_semana == "Sábado"').valor_da_conta
```

```
## DOMINGO
## outra opção: gorjetas[gorjetas.dia_da_semana == 'Domingo'].valor_da_conta

valor_domingo = gorjetas.query('dia_da_semana == "Domingo"').valor_da_conta
```

```
## hora de rodar o ranksums para descobrir o p-value

r2 = ranksums(valor_sabado,valor_domingo)
print(f'O valor de p-value é {r2.pvalue:.2f}')

## p-value > 0,05; portanto, aceitaremos a hipótese nula
## ou seja, a distribuição do valor da conta é a mesma nos dois dias
```

## 2) Teste de hipótese

### Valor da conta de acordo com a hora do dia

### H<sup>null</sup> (Hipótese nula)  p-value > 0,05
A distribuição do valor da conta é a mesma nos dois períodos do dia

### H<sup>alt</sup> (Hipótese alternativa)  p-value < 0,05
A distribuição do valor da conta não é a mesma nos dois períodos do dia

```
## a hora do dia tem influência sobre o valor da conta?
## calculemos a média do valor gasto em cada período do dia

gorjetas.groupby('hora_do_dia').mean()[['valor_da_conta', 'gorjeta', 'porcentagem']]

## nota-se que a média do valor gasto é maior no jantar, assim como a gorjeta
## porém, a média do percentual da gorjeta em relação à conta é menor
```

```
## façamos o teste de hipótese para comparar o comportamento da distribuição dos valores nos dois períodos do dia

r3 = ranksums(valor_almoco, valor_jantar)

## p-value < 0,05; portanto, aceitaremos a hipótese alternativa
## ou seja, a distribuição do valor da conta não é a mesma nos dois períodos do dia
```

```
print(f'O valor de p-value é de {r3.pvalue:.3f}; dessa forma, aceitaremos a hipótese alternativa')
```

### Percentual da gorjeta em relação ao valor da conta de acordo com a hora do dia

### H<sup>null</sup> (Hipótese nula)  p-value > 0,05
A distribuição do percentual de gorjeta é a mesma nos dois períodos do dia

### H<sup>alt</sup> (Hipótese alternativa)  p-value < 0,05
A distribuição do percentual de gorjeta não é a mesma nos dois períodos do dia

```
## montemos duas queries pra separar o percentual de gorjeta da refeição de quem foi almoçar do de quem foi jantar

porcentagem_almoco = gorjetas.query('hora_do_dia == "Almoço"').porcentagem
porcentagem_jantar = gorjetas.query('hora_do_dia == "Jantar"').porcentagem
```

```
## façamos o teste de hipótese para saber o comportamento da distribuição das porcentagens de gorjeta de acordo com a hora do dia

r4 = ranksums(porcentagem_almoco, porcentagem_jantar)
r4

## p-value > 0,05; portanto, aceitaremos a hipótese nula
## ou seja, a distribuição dos percentuais de gorjeta é a mesma nos dois períodos do dia
```

```
print(f'O valor de p-value é de {r4.pvalue:.2f}; dessa forma, aceitaremos a hipótese nula')
```

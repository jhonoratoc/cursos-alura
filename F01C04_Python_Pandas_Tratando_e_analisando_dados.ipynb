# <font color=navyblue> AULA 06 - Tratando dados faltantes

## 1) Tirando valores nulos

```python
## façamos todo o processo novamente para fins didáticos

import pandas as pd
dados_06 = pd.read_csv('aluguel_residencial.csv', sep = ';')
dados_06.head(10)
```

```python
## nota-se que as variáveis Condominio e IPTU possuem dados NaN (not a number)
## devemos tratar esses dados e dar um destino a essas lacunas

## df.isnull() cria um df booleano indicando se o item é nulo ou não

dados_06.isnull()
```

```python
## df.notnull() cria um df booleano indicando o resultado oposto do apresentado em df.isnull()

dados_06.notnull()
```

```python
## vejamos as informações do df

dados_06.info()

## valor, condomínio e iptu possuem valores nulos, visto que possuem menos de 22580 valores não-nulos
```

```python
## vamos verificar quais são as linhas com Valor nulo
## para isso, criaremos uma Series booleana para eliminar os valores False e manter apenas os True
## nesse caso, True = valores nulos; False = valores não nulos

dados_06[dados_06['Valor'].isnull()]
```

```python
## por não servirem para avaliação, queremos eliminar as linhas de Valor = NaN
## para isso, usaremos o comando df.dropna(subset = ['Valor'], inplace = True)


dados_06.dropna(subset = ['Valor'], inplace = True)      ## eliminando
dados_06[dados_06['Valor'].isnull()]                     ## verificando se foi eliminado

## todas as linhas que apresentavam valores nulos na coluna Valor foram eliminados
```

## 2) Tratamento condicional

```python
## trataremos as colunas Condominio e IPTU

## no caso do Condomínio, é de conhecimento geral que casas não pagam condomínio, ao contrário de apartamentos e quitinetes
## não devemos eliminar as casas por terem valores NaN na coluna Condomínio; em vez disso, devemos substituir os dados faltantes
## por valores não nulos

## no caso dos apartamentos com valores NaN na coluna Condomínio, as linhas correspondentes devem ser eliminadas
```

```python
dados_06[dados_06['Condominio'].isnull()]

## 1813 valores NaN na coluna Condominio
```

```python
## façamos uma seleção contendo os valores que nós queremos eliminar
## no caso, apartamentos com valor de condomínio NaN 

selecao = (dados_06['Tipo'] == 'Apartamento') & (dados_06['Condominio'].isnull())

## note que foram selecionados os dados que nós NÃO QUEREMOS no df
## se eu fizer dados_06[selecao], serão eliminados os dados que eu quero manter

## para que isso não aconteça, devemos inverter a seleção (True vira False e vice-versa)
## como fazer isso? basta colocar um til (~) antes de selecao

dados_06 = dados_06[~selecao]
dados_06
```

```python
dados_06[dados_06['Condominio'].isnull()]

## todas as linhas de apartamentos sem dados de condomínio foram eliminadas
```

```python
## agora vamos analisar os casos de casas com valores de condomínio e IPTU NaN
## devemos substituir os valores nulos por valores não-nulos 

dados_06.fillna(value = 0, inplace = True)  ## esse comando substitui TODOS os valores NaN do df

dados_06.index = range(len(dados_06))  ## ajustando o índice

dados_06  ## pronto!
```

```python
## outra forma de fazer:

############### dados_06.fillna(value = {'Condominio': 0, 'IPTU': 0}, inplace = True)

## dessa forma, eu posso escolher em qual coluna substituir os valores (e atribuir valores diferentes para cada caso)
## nem sempre eu quero substituir TODOS os valores NaN do meu df
```

```python
## vamos exportar o novo df para um arquivo csv

dados_06.to_csv('aluguel_residencial.csv', sep = ';', index = False)  ## lembrar do index = False para não duplicar o índice
```

# <font color=navyblue> AULA 07 - Novas variáveis

## 1) Criando novas variáveis

```python
## precisamos de três novas variáveis:
  ## valor bruto / valor bruto por m²
  ## aluguel por m²
  ## identificador de casa/apartamento
```

```python
## vamos fingir que é um novo notebook

import pandas as pd
dados_07 = pd.read_csv('aluguel_residencial.csv', sep = ';')
dados_07.head(10)
```

```python
## criemos a coluna Valor Bruto (somatório de todos os valores a serem pagos por mês)

dados_07['Valor Bruto'] = dados_07['Valor'] + dados_07['Condominio'] + dados_07['IPTU']
dados_07.head(10)

## ✅
```

# <font color=navyblue> AULA 09 - Removendo outliers


## 1) Identificando e removendo outliers

```python
## vamos remover os dados estranhos do nosso dataframe

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6))  ## configuração do tamanho do gráfico
```

```python
## a técnica a ser usada para remover os outliers é o BOXPLOT
## vamos importar os dados

dados_09 = pd.read_csv('aluguel_residencial.csv', sep = ';')
dados_09.head(10)
```

```python
## façamos o BOXPLOT de uma das colunas do df

dados_09.boxplot(column='Valor')

## não dá pra ver direito, o boxplot tá achatado pelos outliers
## outro detalhe é que a visualização na vertical não é tão boa
```

```python
## criemos uma Series com a coluna Valor do df

valor_09 = dados_09['Valor']

## abaixo, um exemplo do boxplot
## os dados estão divididos em quatro quartis, cada um com 25% dos valores, organizados em ordem crescente
## os pontinhos pretos são os outliers, aqueles que queremos eliminar
```

<img src="boxplot-quartiles.png" width=70%>

```python
## separemos os quartis dos nossos dados

Q1 = valor_09.quantile(.25)  ## mostremos o primeiro quartil (25% dos dados)
Q3 = valor_09.quantile(.75)  ## mostremos o terceiro quartil (75% dos dados)

IIQ = Q3-Q1  ## IIQ = intervalo interquartil (esse retângulo maior na foto acima

limite_inf = Q1 - 1.5*IIQ
limite_sup = Q3 + 1.5*IIQ

## os limites determinam onde estarão os outliers (ou seja, determinam o intervalo de dados a ser aproveitado/descartado)
```

```python
selecao_09 = (valor_09 >= limite_inf) & (valor_09 <= limite_sup)
dados_09n = dados_09[selecao_09]

## ok, dados corrigidos e outliers retirados
## vamos refazer o boxplot

dados_09n.boxplot(column = 'Valor')

## como os outliers foram retirados, o boxplot recalculou todos os valores de média, mediana e os intervalos dos quartis
## por consequência, novos valores foram considerados outliers
## para decidir se esses novos outliers devem ser descartados, é necessário fazer uma análise mais aprofundada dos valores
```

```python
## façamos histogramas para comparar os dados antes e depois da retirada dos outliers


dados_09.hist(column = 'Valor')
dados_09n.hist(column = 'Valor')  ## esse histograma tá bem melhor
```

## 2) Identificando e removendo outliers por grupo

```python
## façamos a remoção dos ouliers, agrupando os imóveis pelo tipo
## (Apartamento, Casa, Casa de Condomínio, Casa de Vila, Quitinete)

## o parâmetro by trabalha como o groupby

dados_09.boxplot(column = 'Valor', by = 'Tipo')

## separando os dados por grupo, a análise torna-se mais completa
## certos valores, que são esperados para um tipo específico de imóvel, podem não ser adequados a outros tipos
```

```python
## vamos montar um groupby pra calcular os parâmetros do boxplot por tipo de imóvel
## filtraremos apenas a Series 'Valor'

grupo_tipo09 = dados_09.groupby(by = 'Tipo')['Valor']
grupo_tipo09  ## SeriesGroupBy
```

```python
## calcularemos os parâmetros do boxplot usando os dados dessa SeriesGroupBy


Q1 = grupo_tipo09.quantile(.25)  ## mostremos o primeiro quartil (25% dos dados)
Q3 = grupo_tipo09.quantile(.75)  ## mostremos o terceiro quartil (75% dos dados)

IIQ = Q3-Q1  ## IIQ = intervalo interquartil (esse retângulo maior na foto acima

limite_inf = Q1 - 1.5*IIQ
limite_sup = Q3 + 1.5*IIQ
```

```python
## dados os parâmetros, hora de fazer a seleção para retirar os outliers

dados_09new = pd.DataFrame()  ## df zerado
for tipo in dados_09['Tipo'].unique():  ## para cada tipo de imóvel contido nesse array
    bool_tipo = dados_09['Tipo'] == tipo  ## booleano -> tipo de imóvel
    bool_limites = (dados_09['Valor'] >= limite_inf[tipo]) & (dados_09['Valor'] <= limite_sup[tipo])  ## booleano -> limites bp
    selecao = bool_tipo & bool_limites  ## unindo os booleanos
    dados_09selecao = dados_09[selecao]  ## aplicando o booleano ao df
    dados_09new = pd.concat([dados_09new, dados_09selecao])  ## concatenação

    
```
